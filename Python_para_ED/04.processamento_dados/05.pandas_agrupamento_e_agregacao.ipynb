```python
import pandas as pd
```

```python
dados = pd.read_csv('./dados/users.csv')
dados.head()
```

```python
# vamos trabalhar com 2 novos conceitos: AGRUPAMENTO e AGREGAÇÃO

# AGRUPAMENTO: é o nosso conhecido group by, que concatena valores coincidentes em uma coluna informada
# AGREGAÇÃO: agrupa os valores de uma coluna informada, por meio de uma operação matemática informada
```

```python
# exemplo: vamos agrupar os usuários por profissão e calcular a média das idades de cada profissão

dados_profissao = (
    dados
    .groupby('profession')
    .agg(
        media_idades=('age', 'mean')
        )
)
dados_profissao
```

```python
# agora vamos brincar com a coluna de idade

dados_profissao = (
    dados
    .groupby('profession')
    .agg(
        idade_min=('age','min'),
        idade_max=('age', 'max'),
        media_idades=('age', 'mean')
        )
)
dados_profissao
```
