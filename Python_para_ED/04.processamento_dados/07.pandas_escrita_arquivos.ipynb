```python
import pandas as pd
```

```python
# vamos primeiro abrir o df

dados = pd.read_csv('./dados/users.csv')
```

```python
# agora, vamos fazer um pequeno processamento, criando uma coluna com o log de criação

from datetime import datetime as dt

dados['log_criacao'] = dt.now()
dados
```

```python
# vamos, agora, salvar esse novo df dentro da nossa camada raw

dados.to_csv('./dados/raw/csv/dados_log.csv', index=False)
```

```python
# por fim, vamos salvar o novo dataframe, mas agora em json

dados.to_json('./dados/raw/json/dados_log.json', orient='records')
```
