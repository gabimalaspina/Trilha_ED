```python
import pandas as pd
```

```python
dados = pd. read_csv('./dados/users.csv')
dados
```

```python
 # FILTRO POR COLUNA: vamos agora criar um novo dataframe contendo apenas as colunas de nosso interesse

dados_filtrado = dados[['id', 'firstname', 'email', 'profession']]
dados_filtrado
```

```python
# FILTRO POR VALORES DE UMA COLUNA = sempre usamos chaves duplas [ [ ] ]

dados_acima_30 = dados[ dados['age'] > 30 ]
dados_acima_30
```

```python
# agora, vamos filtrar com base em valores de múltiplas colunas utilizando o operador lógico AND >>> &
# lembrando sempre de utilizar df[   df[ ]   ]; nesse caso, teremos df[    (df['coluna1'] == 'condição 1') & (df['coluna2']=='condição2)]

dados_30_bombeiros = dados[ ( dados['age'] > 30 ) & ( dados['profession'] == 'firefighter' ) ]
dados_30_bombeiros
```

```python
# para facilitar a leitura, podemos usar parênteses e identação a nosso favor:

dados_30_bombeiros = (
    dados[ # chamando as colunas que atendam às condições:
        (dados['age'] > 30) # condição 1
        & # E
        (dados['profession'] == 'firefighter') # condição 2
    ]
)
dados_30_bombeiros
```

```python
# agora, vamos utilizar a mesma lógica para procurar emails com .br na coluna email OU email2 >>> OU = |

dados_comp_email = (
    dados[ # chamando as colunas que atendam às condições:
        (dados['email'].str.contains('.com.br')) # condição 1
        | #OU
        (dados['email2'].str.contains('.com.br')) # condição 2
    ]
)
dados_comp_email
```
