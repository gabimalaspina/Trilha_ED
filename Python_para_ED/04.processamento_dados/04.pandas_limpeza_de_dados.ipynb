```python
import pandas as pd
```

```python
dados = pd.read_csv('./dados/users.csv')
dados.describe()
```

```python
# podemos ver, após executar describe, que na coluna idade há 2 valores nulos - NaN

# primeiro, então, vamos localizá-los

dados_idade_NaN = (
    dados[
        dados['age'].isnull()
    ]
)
```

```python
dados_idade_NaN # NaN = "not a number"
```

```python
# uma opção é eliminar esses valores utilizando uma função específica do pandas: dropna

dados_drop_NaN = dados.dropna(subset=['age'])
```

```python
print(dados_drop_NaN)
dados_drop_NaN.describe()
```

```python
# uma outra opção seria preencher os campos com NaN com um valor , para não perder os dados, utilizando a função fillna

# nesse caso, para as idades faltantes, vamos preencher esses campos com a média das idades

# é claro que, na vida real, vai depender totalmente do contexto e das regras de negócio, essa tratativa utilizando a média é só um exemplo

dados_fill_NaN = dados.fillna(
    {
        'age': round(dados['age'].mean(), 0)
    }
)
```

```python
dados_fill_NaN
```

```python
# outro processo comum nessa etapa de limpeza é RENOMEAR as 

dados_renomeado = dados.rename(columns={'name':'firstname', 'email2':'2nd email'})
dados_renomeado
```
