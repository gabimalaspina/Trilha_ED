```python
import pandas as pd
```

```python
# vamos juntar informações de 2 diferentes dataframes

# primeiro, criando os dataframes a partir dos diferentes arquivos:

dados_usuarios = pd.read_csv('./dados/users.csv')

dados_status = pd.read_csv('./dados/users_status.csv')
```

```python
# há 2 formas de fazer a junção: utilizando a função merge e utilizando a função join
# porém, a função merge é mais "flexível" e se assemelha mais ao join do sql, então acaba sendo mais utilizada

dados_inner_join = pd.merge(
    dados_usuarios,
    dados_status,
    on = 'id',
    how = 'inner'
)
dados_inner_join
```
