```python
dados = pd.read_csv('./dados/users.csv')
dados
```

```python
# agora, vamos acrescentar uma nova coluna

dados['nova_coluna']='valor_desejado'
```

```python
# agora, vamos criar um campo calculado

dados['idade_soma_10']=dados['age']+10
```

```python
# vamos agora juntar primeiro nome e sobrenome

dados['nome_completo']=dados['firstname']+' '+dados['lastname']
```

```python
# agora trazendo para um exemplo mais real na vivência do engenheiro, podemos precisar criar colunas com o "log" de criação
# para isso, utilizamos o módulo datetime

from datetime import datetime as dt
```

```python
# para criar uma coluna com o momento do registro

dados['log_criacao']=dt.now()
```

```python
dados
```

```python
# agora, para adicionar apenas a data formatada

dados['log_criacao_data']=dt.today().strftime('%d/%m/%Y')
```

```python
# e, por fim, podemos precisar eliminar colunas - utilizamos o método drop, informando as colunas que queremos deletar

dados.drop(columns=['email2', 'nova_coluna', 'idade_soma_10'])
```
